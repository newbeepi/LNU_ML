```python
import matplotlib.pyplot as plt
```

```python
def f(x):
    return (3-x)**2

def df(x):
    eps = 10e-11
    return (f(x+eps) - f(x))/eps
```

```python
plt.plot(range(7), [f(x) for x in range(7)])
```

```python
from scipy.misc import derivative
```

```python
dy_custom = df(0)
dy_lib = derivative(f, 0)
print("custom dy: ", dy_custom)
print("lib dy: ", dy_lib)
```

```python
def gradient_descent(alpha, eps, epoch, df=df):
    x_old = 0
    y_old = f(x_old)
    
    points = {x_old: y_old}
    for i in range(epoch):
        x_new = x_old - alpha*df(x_old)
        y_new = f(x_new)
    
        points[x_new] = y_new
        
        if abs(x_new-x_old) < eps:
            break
        x_old = x_new
    return points
```

```python
for i in range(5):
    eps = 10 ** (-1 * ((i+1) * 2))
    for i in range(20):
        alpha = 0.05 * (i+1)
        print("alpha=", alpha, "eps=", eps)
        gradient_with_custom_function = gradient_descent(alpha, eps, 10000)
        print("iterations with custom function: ", len(gradient_with_custom_function))

        lib_func = lambda x: derivative(f, x)
        gradient_with_lib_function = gradient_descent(alpha, eps, 10000, df=lib_func)
        print("iteration with lib function: ", len(gradient_with_lib_function))
        print()
```

```python
X_f = range(-2, 5)
Y_f = [f(x) for x in X_f]
plt.plot(X_f, Y_f)

X_gradient = list(gradient_with_custom_function.keys())
Y_gradient = list(gradient_with_custom_function.values())
plt.scatter(X_gradient, Y_gradient)
plt.plot(X_gradient[-1], Y_gradient[-1], marker='o', markersize=10)
```
